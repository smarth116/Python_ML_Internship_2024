```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
sm
```

```python
sm['Percentage']=sm['Total']/5
```

```python
sm
```

```python

def determine_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B+'
    elif percentage >= 60:
        return 'B'
    elif percentage >= 50:
        return 'C'
    elif percentage >= 35:
        return 'D'
    else:
        return 'F'

sm['Grade'] = sm['Percentage'].apply(determine_grade)
sm
```

```python
Temp=sm.head()
Temp
```

```python
colors = {
    'English': 'red',
    'Hindi': 'blue',
    'Maths': 'green',
    'Science': 'purple',
    'SST': 'orange'
}
for sub in ['English','Hindi','Maths','Science','SST']:
    plt.figure(figsize=(10,6))
    plt.plot(Temp['Name'],Temp[sub],marker='o',label=sub,color=colors[sub])
    plt.title("Marks in different Subjects")
    plt.xlabel('Name')
    plt.ylabel('Marks')
    plt.legend()
    plt.grid(True)
    plt.show()
```
