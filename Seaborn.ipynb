```python
auto =pd.read_csv('Automobile_data.csv')
auto
auto['price'] = pd.to_numeric(auto['price'], errors='coerce')
def determine_type (price):
    if price < 6000:
        return 'under budget'
    elif price <=8000 :
        return 'afford able'
    elif price <=10000 :
        return 'little bit expensive but i can afford'
    elif price <= 15000:
        return 'over budget'
    elif price <= 25000:
        return 'luxury'
    else :
        return 'most expensive'
  
budget_categories = []
for price in auto['price']:
    budget_categories.append(determine_type(price))
auto['budget'] = budget_categories
print(auto)
```
