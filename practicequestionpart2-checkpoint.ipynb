```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
```

```python
news=pd.read_csv('titanic.csv')
news
```

```python
classvsalive=news.groupby (['alive','class']).size().unstack(fill_value=0)
classvsalive.plot(kind='line',stacked=True)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=news["survived"],linewidth=1.5,label="Alone")
sns.lineplot(data=news["pclass"],linewidth=1.5,label="Survived")
plt.title("Alone")
plt.ylabel("MALE FEMALE")
plt.xlabel("Survived")
plt.show()
```

```python
Age=news.groupby (['sex','age']).size().unstack(fill_value=0)
Age.plot(kind='bar',stacked=True)
plt.show()
```

```python
which_town=news.groupby (['who','embark_town']).size().unstack(fill_value=0)
which_town.plot(kind='bar',stacked=True)
plt.show()
```
